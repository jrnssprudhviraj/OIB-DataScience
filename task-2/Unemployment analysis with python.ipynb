```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load the data
data = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

# Trim column names to remove leading or trailing spaces
data.columns = data.columns.str.strip()

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# # Set the 'Date' column as the index
# data.set_index('Date', inplace=True)

# # Reindex with monthly frequency
# data.index = pd.date_range(start=data.index.min(), end=data.index.max(), freq='MS')




# Plot the time series data
def plot_time_series(data, ylabel):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data, color='blue')
    plt.title('Time Series Plot')
    plt.xlabel('Date')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

plot_time_series(data['Estimated Unemployment Rate (%)'], 'Unemployment Rate (%)')

# Fit ARIMA model
def fit_arima_model(data, order):
    model = ARIMA(data, order=order)
    model_fit = model.fit()
    return model_fit

order = (5, 1, 0)  # Example order
model_fit = fit_arima_model(data['Estimated Unemployment Rate (%)'], order)
print(model_fit.summary())

# Generate forecasts
def forecast_arima_model(model_fit, steps):
    forecast = model_fit.forecast(steps=steps)
    return forecast

forecast_steps = 12  # Example: forecast for the next 12 months
forecast = forecast_arima_model(model_fit, forecast_steps)
print("Forecasted values:")
print(forecast)

# Plot actual vs forecasted values
def plot_forecast(actual_data, forecast, ylabel):
    plt.figure(figsize=(12, 6))
    plt.plot(actual_data.index, actual_data, label='Actual', color='blue')
    plt.plot(forecast.index, forecast, label='Forecast', color='red')
    plt.title('Actual vs Forecasted')
    plt.xlabel('Date')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_forecast(data['Estimated Unemployment Rate (%)'], forecast, 'Unemployment Rate (%)')
```
